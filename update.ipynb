```python
import pandas as pd
df=pd.read_csv("futuredatest2.csv")
```

```python
row,column = df.shape
pastdate=str(df.iat[row-1,2])
```

```python
import datetime
startday = datetime.datetime.strptime(pastdate, '%Y%m%d')
k=[]
k.append(startday.strftime("%Y%m%d"))
i = datetime.datetime.now()
a=str(i)
now=a[:a.index(" ")]
startday = datetime.datetime.strptime(now, '%Y-%m-%d')
k.append(startday.strftime("%Y%m%d"))
```

```python
import requests
import codecs
import xml.etree.cElementTree as ET
if '0' in k[1][4:6]:
    month = k[1][4]
else:
    month = k[1][4:6]
if '0' in k[0][4:6]:
    month = k[0][4]
else:
    month = k[0][4:6]
a=datetime.datetime(int(k[1][:4]),int(k[1][4:6]),int(k[1][6:8]))
b=datetime.datetime(int(k[0][:4]),int(k[0][4:6]),int(k[0][6:8]))
timedelta=(a-b).days+1
def dateprepare(pastdate,daylen):
    startday = datetime.datetime.strptime(pastdate, '%Y%m%d')
    k=[]
    k.append(startday.strftime("%Y%m/%d"))
    for i in range(daylen):
        newday = (startday+datetime.timedelta(days=1))
        formatday=newday.strftime("%Y%m/%d")
        k.append(formatday)
        startday=newday
    return k
def change(filecontent):
    with open("data.xml",'wb+') as f:
        f.write(filecontent)
    with codecs.open("data.xml",'r',encoding='gbk') as fp:
        text = fp.read().replace('<?xml version="1.0" encoding="GBK"?>', '<?xml version="1.0" encoding="UTF-8"?>')
    return text
def getcontent(current,timedelta):
    base="http://www.cffex.com.cn/sj/hqsj/rtj/"
    headers = {
        'Host': 'www.cffex.com.cn',
        'Referer': 'http://www.cffex.com.cn/rtj/',
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
        'X-Requested-With': 'XMLHttpRequest',
    }
    whole=[]
    timelist=dateprepare(current,timedelta)
    for i in range(1,timedelta):
            url=base+timelist[i]+"/index.xml"
            response = requests.get(url, headers=headers)
            if response.status_code == 200:
                b=[]
                xmlcontent=response.content
                verify="http://www.w3.org/TR/xhtml1/DTD/xhtml1-transitional.dtd"
                if verify not in str(xmlcontent):      
                    handlecontent=change(xmlcontent)
               # print(handlecontent)
                    root = ET.fromstring(handlecontent)
                    for j in range(5):
                        if "IF" in root[j][0].text:
                            for i in range(6):
                                b.append(root[j][i].text.strip(' '))
                            break
            if len(b)!=0:
                whole.append(b)
                with open('wholelist.txt','w+') as file:
                    for i in b:
                        file.write(i)
                    file.write('\n')
    return whole
a=getcontent(k[0],timedelta)
```
